```python
import os
import re
import numpy as np
from tqdm import tqdm

import tensorflow as tf
from transformers import *
```

```python
eng_sentiment_model = TFBertClassifier_ENG(model_name='bert-base-cased')
```

```python
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.metrics import CategoricalAccuracy

optimizer = tf.keras.optimizers.legacy.Adam(
    learning_rate=5e-05, # this learning rate is for bert model , taken from huggingface website 
    epsilon=1e-08,
    decay=0.01,
    clipnorm=1.0)

# Set loss and metrics
loss = CategoricalCrossentropy(from_logits = True)
metric = CategoricalAccuracy('balanced_accuracy'),
# Compile the model
eng_sentiment_model.compile(
    optimizer = optimizer,
    loss = loss, 
    metrics = metric)
```

```python
eng_sentiment_model.load_weights('english_multi_sentiment_weights')
```

```python
test = []
adj_test = []

filter = re.compile('[a-zA-Z0-9]+')
for i in kitsch:
    if filter.findall(i):
        test.append(" ".join(filter.findall(i)))

for j in test:
    try:
        if detect(j) == 'en':
            adj_test.append(j)
    except:
        continue

adj_test
```

```python
result = []

for i in adj_test:
    result.append(eng_emotion_predict(i, eng_sentiment_model))

result
```
